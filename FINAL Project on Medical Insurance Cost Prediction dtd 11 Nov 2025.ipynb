```python
# dervied feature from BMI value
def BMI_Category(value):
    if value < 18:
        return 'Underweight'
    elif 18<= value < 25:
        return 'Normal Weight'
    elif value >= 25:
        return 'Over weight'
    else:
        return 'Obese'

df["BMI_Category"] = df['bmi'].apply(BMI_Category)
print(df)
```

```python
#Univariate
sns.boxplot(x=df["medication_count"], color='yellow')
plt.title("Boxplot for Insurance Charges")
plt.xlabel("Medication_count")
plt.show()
```

#### INSIGHTS

```python
#univariate plot
plt.figure(figsize=(8,6))
sns.histplot(df['age'], bins=15, kde=True, color='violet')
plt.title('Age Distribution')
plt.show()
```

### INSIGHTS

•As age increases, health risks rise — leading to higher predicted medical insurance costs.
•Insurance cost tends to increase with age and contribute to higher average medical expenses.

```python
## bivariate with another variables add on
sns.scatterplot(x='bmi', y='age', data=df, color='purple')
plt.title("BMI vs age")
plt.xlabel("BMI")
plt.ylabel("Age")
plt.show()
```

### INSIGHTS

• Almost all age groups greatly have significance with BMI and larger the BMI value, larger will be the insurance cost.

```python
# using hue bivariate
plt.figure(figsize=(6,4))
sns.barplot(x='age', y='sex', hue='smoker', data=df, palette='Set1')
plt.title('Age vs Sex')
plt.show()
```

### INSIGHTS

•Smoking habits are evenly distributed across different age and sex groups.

```python
#multivariate- more data comparison -another plot
sns.pairplot(df[['age', 'bmi', 'diabetes']], hue='diabetes', palette='Set2')
plt.suptitle("Pair plot for Age Vs Diabetes", y=1.02)
plt.show()
```

### INSIGHTS

```python
•The pair plot shows that individuals with diabetes tend to be older and have higher BMI compared to non-diabetic individuals.
While age and BMI are not strongly correlated with each other, both seem to have an individual influence on diabetes risk.
```

```python
#multivariate
corr = df[['age', 'policy_term_years', 'annual_premium']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Heatmap showing Age, Policy Term Years, and Annual Premium')
plt.show()
```

### INSIGHTS

```python
• The age directly depends on the annual premium, Higher the age of the individual, greater will be the premium.
```
