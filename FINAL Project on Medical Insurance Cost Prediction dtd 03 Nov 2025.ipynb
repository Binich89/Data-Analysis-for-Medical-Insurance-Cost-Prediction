```python
# dervied feature from BMI value
df["BMI_Category"] = df["bmi"].apply(lambda x: "Underweight" if x < 18.5 
                                     else "Normal" if x < 25 
                                     else "Overweight" if x < 30 
                                     else "Obese")
```

```python
print(df[["age", "bmi", "BMI_Category"]].head())
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['bmi'], color='lightgreen')
plt.title('BMI Distribution')
plt.show()
```

#### INSIGHTS

```python
plt.figure(figsize=(6,4))
sns.histplot(df['age'], bins=10, kde=True, color='skyblue')
plt.title('Age Distribution')
plt.show()
```

### INSIGHTS

•As age increases, health risks rise — leading to higher predicted medical insurance costs.
•Insurance cost tends to increase with age and contribute to higher average medical expenses.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='bmi', y='marital_status', data=df, color='lightgreen')
plt.title('BMI vs Marital Status')
plt.show()
```

### INSIGHTS

•Almost all marital groups have similar BMI, indicating that marital status has little effect on insurance cost.

```python
plt.figure(figsize=(6,4))
sns.barplot(x='region', y='sex', data=df, palette='Set2')
plt.title('Region vs Sex')
plt.show()
```

### INSIGHTS

•The gender distribution is uniform across regions, region or gender alone does not strongly influence insurance cost and slight importance in the increase of insurance cost.

```python
sns.pairplot(df[['age', 'bmi', 'sex']], diag_kind='kde', corner=True)
plt.suptitle("Pairwise Relationships", y=1.02)
plt.show()
```

### INSIGHTS

```python
•The combined effect on insurance cost is additive rather than dependent — meaning both factors age and bmi independently contribute to higher costs.
```

```python
num_df = df.select_dtypes(include=['int64', 'float64'])
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

### INSIGHTS

```python
• The insurance cost is mainly driven by medical and health-related factors such as average claim amount, monthly premium, deductible, and hospitalization days,
while demographic variables like income or dependents have little effect.
```

```python
num_cols= ['age', 'bmi', 'income', 'risk_score']
```

```python
sns.pairplot(df[num_cols], diag_kind='kde', corner=True)
plt.suptitle("Pairwise Relationships Among Key Variables", y=1.02, fontsize=14, fontweight='bold')
plt.show()
```

### INSIGHTS

```python
•The risk score strongly influences insurance cost and increases with age and slightly with BMI, but shows little dependence on income.
•Older and overweight individuals are more likely to incur higher medical insurance costs.
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='region', y='annual_medical_cost', hue='smoker', data=df, palette='pastel')
plt.title("Region vs Annual Medical Cost by Smoker Status", fontsize=14, fontweight='bold')
plt.show()
```

### INSIGHTS

```python
• Smokers incur higher annual medical costs, indicating a strong positive relationship between smoking and insurance cost.
• Former and non-smokers have lower and more stable medical expenses, suggesting that smoking is one of the strongest predictors
    of high insurance charges.
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='age', y='sex', hue='smoker', data=df)
plt.title("Age vs Sex by Smoker Status")
plt.show()
```

### INSIGHTS

```python
• Smokers of all ages and genders states that increase in health risks are likely to face higher medical insurance costs regardless of gender.
```

### INSIGHTS
